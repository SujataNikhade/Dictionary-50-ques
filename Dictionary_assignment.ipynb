```python
##1. Sort Dictionary by Value Write a Python script to sort (ascending and descending) a dictionary by value.
def get_value(item):
    return item[1]
def sorted_dict(dic,descending=False):
    return dict(sorted(dic.items(),key=get_value,reverse=descending))
my_dict={"A":10,
         "B":3,
        "C":2,
        "D":11}
print("Ascending:",sorted_dict(my_dict))
print("Decending:",sorted_dict(my_dict,descending=True))
```

```python
#Concatenate Dictionaries Write a Python script to concatenate the following dictionaries to create a new one.
def concatenate_dict(dic1,dic2,dic3):
    result={}
    result.update(dic1)
    result.update(dic2)
    result.update(dic3)
    return result
dic1 = {1: 10, 2: 20}
dic2 = {3: 30, 4: 40}
dic3 = {5: 50, 6: 60}
concatenate_dict(dic1,dic2,dic3)
```

```python
##Merge Two Python Dictionaries Write a Python script to merge two Python dictionaries.
def merge_dicts(d1,d2):
    merge={}
    for i in(d1,d2):
        for key,value in i.items():
            merge[key]=value
    return merge
d1={1:10,2:20}
d2={3:30,4:40}
merge_dicts(d1,d2)
```

```python
#Split Dictionary of Lists into List of Dictionaries
data = {'Science': [88, 89, 62, 95], 'Language': [77, 78, 84, 80]}
keys = list(data.keys())
values = list(data.values())
result = []
for i in range(len(values[0])):
    new_dict = {}
    for k in keys:
        new_dict[k] = data[k][i]
    result.append(new_dict)
print("Split said dictionary of lists into list of dictionaries:")
print(result)
```

```python
#49.Write a Python program to convert string values of a given dictionary into integer/float datatypes
data = [{'x': '10.12', 'y': '20.23', 'z': '30'},
        {'p': '40.00', 'q': '50.19', 'r': '60.99'}]
result = []
for d in data:
    new_dict = {}
    for k, v in d.items():
        new_dict[k] = float(v)
    result.append(new_dict)

print("\nString values converted to float types:")
print(result)
```
